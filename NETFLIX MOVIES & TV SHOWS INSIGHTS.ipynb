## **DATA CLEANING**

**FINDING MISSING VALUES**

```python
df.isnull().sum()
```

**FILLING MISSING VALUES**

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Not Mentioned')
df['country'] = df['country'].fillna('Not Specified')
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['duration'] = df['duration'].fillna('Not Available')
```

```python
df.isnull().sum()
```

**REMOVE DUPLICATES**

```python
df.drop_duplicates(inplace=True)
```

**CORRECTING DATE TYPES**

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

**DURATION COLUMN DIVIDED INTO duration_type AND duration_value**

```python
df['duration_type'] = df['duration'].apply(lambda x: str(x).split()[-1] if x != 'Not Available' else 'Unknown')
df['duration_value'] = df['duration'].apply(lambda x: str(x).split()[0] if x != 'Not Available' else None)

df['duration_value'] = pd.to_numeric(df['duration_value'], errors='coerce')
```

```python
df[['duration', 'duration_type', 'duration_value']].head()
```

**CREATING DERIVED COLUMNS**

```python
#checking which year added
df['year_added'] = df['date_added'].dt.year

#find whether Netflix adds older or newer releases more often
df['content_age'] = df['year_added'] - df['release_year']
```

#### **UNIVARIATE ANALYSIS**

**1.Content type distribution**

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type')
plt.title('Distribution of Content Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

print(df['type'].value_counts())
```

**Insights:**

Movies dominate Netflix’s library.  
TV Shows make up a smaller but significant share.  
This shows Netflix’s focus on film content while expanding series steadily.

**2.Rating Distribution**

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index, color='brown')
plt.title('Distribution of Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

print(df['rating'].value_counts())
```

**Insights:**

TV-MA and TV-14 are the most common ratings.  
Mature audience content dominates Netflix’s library.  
Family or kids’ ratings (like “TV-Y”) are less frequent.

**3.Release Year Trend**

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='release_year', color='skyblue')
plt.title('Number of Titles Released per Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

print(df['release_year'].value_counts().sort_index())
```

**Insights:**

  Content production surged after 2015.  
Peak release years are around 2018 to 2020.  
Indicates Netflix major growth during the streaming boom.

**4.Duration Distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['duration_value'], bins=20, kde=True, color='pink')
plt.title('Distribution of Duration (Movies / Seasons)')
plt.xlabel('Duration Value')
plt.ylabel('Frequency')
plt.show()

```

**Insights**

Most movies last between 80–120 minutes.  
A few outliers show much longer durations.  
TV Shows typically appear as smaller duration values.

**2.Year, Type, and Duration Relationship**

## **VISUALIZATIONS**

**1.Bar Plot – Count of Movies vs TV Shows**

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type')
plt.title("Count of Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

**Insights**

Most of the titles on Netflix are Movies.  
TV Shows are fewer but steadily increasing.  
This shows Netflix mainly focuses on movie content.

**2.Line Chart – Number of Releases Over Years**

```python
year_count = df.groupby('release_year').size()
year_count.plot(kind='line', figsize=(6,4), color='blue', marker='o')
plt.title("Number of Releases Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

**Insights**

The number of releases increased a lot after 2015.  
A sharp rise can be seen between 2018 and 2020.  
This means Netflix has been expanding its content library every year.

**3.Pie Chart – Distribution of Content Type**

```python

type_counts = df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors=['YELLOW','GREEN'])
plt.title("Content Type Distribution")
plt.show()
```

**Insights**

Most titles are movies, fewer are TV shows.  
Netflix mainly focuses on movie content.  
TV shows are slowly increasing in number.

**4.Histogram – Duration of Movies**

```python

plt.figure(figsize=(6,4))
sns.histplot(df[df['type']=='Movie']['duration_value'], bins=20, color='RED')
plt.title("Movie Duration Distribution")
plt.xlabel("Duration (minutes)")
plt.ylabel("Count")
plt.show()
```

**Insights**

The average movie duration is about 1.5 to 2 hours.  
Only a few movies go beyond 2 hours.  
Most Netflix movies are of standard length, suitable for casual viewing.

**5.Box Plot – Duration by Type**

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='type', y='duration_value')
plt.title("Duration Distribution by Type")
plt.xlabel("Type")
plt.ylabel("Duration (minutes or seasons)")
plt.show()
```

**Insights**

Movies: consistent median around 90–120 mins.  
TV Shows: wider range, especially for mini-series vs long-running series.  
Outliers: Very long movies or very short TV episodes.
